```python
import akshare as ak
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

def get_stock_data(column_name='收盘'):
    # 获取所有A股股票代码列表
    stock_list = ak.stock_zh_a_spot_em()['代码'].tolist()

    # 初始化一个空的DataFrame来存储所有股票的数据
    all_stocks = pd.DataFrame(index=ak.stock_zh_a_hist(symbol='000001',period="daily", start_date="20100101", end_date='20231231', adjust="hfq")['日期'])

    # 使用 tqdm 创建一个进度条
    for code in tqdm(stock_list, desc='Downloading'):
        try:
            # 获取单只股票的历史数据
            stock_data = ak.stock_zh_a_hist(symbol=code,period="daily", start_date="20100101", end_date='20231231', adjust="hfq")

            # 检查所需的列是否在数据中
            if column_name in stock_data.columns:
                # 将选定的列添加到大的DataFrame中
                all_stocks[code] = stock_data.set_index('日期')[column_name]
        except Exception as e:
            print(f"Error fetching data for stock: {code}, error: {e}")
            continue

    return all_stocks

# 使用函数
stock_close_prices = get_stock_data('收盘')
stock_close_prices.index = pd.to_datetime(stock_close_prices.index)
```
